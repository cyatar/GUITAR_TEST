```python
import tensorflow as tf
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split



buffersize = 100

print(tf.version.VERSION)
data_E_processed = []
data_B_processed = []
data_Noise_processed = []
```

```python
# 分隔/处理数据 
data_E_origin  = []
with open('./data20231010/E_1.txt', 'r') as f:
    for line in f:
    # 用空格分割每一行的字符串
        words = line.split()
        # 遍历每一个字符串
        for word in words:
            # 尝试把字符串转换为浮点数
            try:
                num = float(word)
                # 如果成功，就把浮点数添加到数组中
                data_E_origin.append(num)
            # 如果失败，就跳过这个字符串
            except ValueError:
                pass

data_B_origin = []
with open('./20231010/B_1.txt', 'r') as f:
    for line in f:
    # 用空格分割每一行的字符串
        words = line.split()
        # 遍历每一个字符串
        for word in words:
            # 尝试把字符串转换为浮点数
            try:
                num = float(word)
                # 如果成功，就把浮点数添加到数组中
                data_B_origin.append(num)
            # 如果失败，就跳过这个字符串
            except ValueError:
                pass


data_Noise_origin  = []
with open('./20231010/Noise_1.txt', 'r') as f:
    for line in f:
    # 用空格分割每一行的字符串
        words = line.split()
        # 遍历每一个字符串
        for word in words:
            # 尝试把字符串转换为浮点数
            try:
                num = float(word)
                # 如果成功，就把浮点数添加到数组中
                data_Noise_origin.append(num)
            # 如果失败，就跳过这个字符串
            except ValueError:
                pass







print(len(data_E_origin )/buffersize)
print(len(data_B_origin )/buffersize)
print(len(data_Noise_origin)/buffersize)




```

```python
data_B_origin
```

```python

data_E_processed = []
data_B_processed = []
data_Noise_processed = []


for i in range(0, len(data_E_origin), buffersize* 3):

        data = data_E_origin[i:i+buffersize * 3]

        if(len(data) < buffersize * 3):
            break
        
        x = data[:buffersize]
        y = data[buffersize:buffersize*2]
        z = data[buffersize*2:]

       
        
        result = np.stack([x, y, z], axis=1)
        data_E_processed.append(result)
     

for i in range(0, len(data_B_origin), buffersize * 3):
            
        data = data_B_origin[i:i+buffersize*3]

        if(len(data) < buffersize * 3):
            break

        x = data[:buffersize]
        y = data[buffersize:buffersize*2]
        z = data[buffersize*2:]
        

        result = np.stack([x, y, z], axis=1)
        data_B_processed.append(result)
       

for i in range(0, len(data_Noise_origin), buffersize * 3):
      
        data = data_Noise_origin[i:i+buffersize * 3]

        if(len(data) < buffersize * 3):
            break

        
        x = data[:buffersize]
        y = data[buffersize:buffersize*2 ]
        z = data[buffersize*2:]

        
        result = np.stack([x, y, z], axis=1)
        data_Noise_processed.append(result)    
        
```

```python
print(len(data_E_processed))
print(len(data_B_processed))
print(len(data_Noise_processed))
```

```python

labels = ['E', 'B', 'Noise']

data_E_processed = np.array(data_E_processed)

# [0]


data_E_Lable = np.zeros(len(data_E_processed))
data_B_processed = np.array(data_B_processed)

# [1]
data_B_Lable = np.ones(len(data_B_processed))

data_Noise_processed = np.array(data_Noise_processed)

# [2]
data_Noise_Lable = np.ones(len(data_Noise_processed))*2

# normalize
data_E_processed_normed = data_E_processed/ 16000

data_B_processed_normed = data_B_processed/ 16000
data_Noise_processed_normed = data_Noise_processed/ 16000

# combine


data_all = np.concatenate((data_E_processed_normed, data_B_processed_normed, data_Noise_processed_normed), axis=0)
data_label = np.concatenate((data_E_Lable, data_B_Lable, data_Noise_Lable), axis=0)
```

```python

%matplotlib inline

# Plot some captures

plt.figure(figsize=(20, 5))


   
plt.plot(data_E_processed_normed[2][:100,0], color='r')
plt.plot(data_E_processed_normed[2][:100,1], color='g')
plt.plot(data_E_processed_normed[2][:100,2], color='b')
    




```

```python

x_train, x_test, y_train, y_test = train_test_split(
    data_all,data_label , test_size=0.3)


print(x_train.shape)
print(x_train)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

model = tf.keras.models.Sequential([
  tf.keras.layers.Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=(100, 3)),
  tf.keras.layers.Conv1D(filters=8, kernel_size=3, activation='relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=30)
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)

print("Test loss:", test_loss)
print("Test acc:", test_acc)
model.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

Y_pred = model.predict(x_test)
y_pred = np.argmax(Y_pred, axis=1)
confusion_matrix = tf.math.confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(confusion_matrix,
            annot=True,
            xticklabels=labels,
            yticklabels=labels,
            cmap=plt.cm.Blues,
            fmt='d', cbar=False)
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# Save the model into an HDF5 file ‘model.h5’
model.save('model.h5')
```
