```python


#TODO: 清洗数据
#TODO：数据插值
#TODO：数据归一化
#TODO：数据切分
#TODO：模型训练

buffersize = 200

path ="./data20231012"

files= os.listdir(path)


def get_data(path):
    files= os.listdir(path)
    
    data_output = []
    for file in files:
        data_process = []
        if not os.path.isdir(file):
            print(file)
            f = open(path+"/"+file)
            iter_f = iter(f)
            for line in iter_f:
                line = line.replace("\n","")
                data_process.append(line)
        data_output.append(data_process)
    return data_output
        
# 数据清洗

def clean_data_float(data):
    for i in range(0,len(data)):
        for j in range(0,len(data[i])):
            
            try:
                data[i][j] = float(data[i][j])
            except:
                print("error")
                print(data[i][j])
                data[i][j] = np.nan
    return data     

def interpolate_data(data):
    temp = []
    interpolated_data=[]
    for i in range(0,len(data)):
        temp = pd.DataFrame(data[i])

        #什么玩意，屁用没有
        temp = temp.interpolate()
        
        #temp = temp.dropna()
        #or temp = temp.fillna(0)?
        
        temp = np.array(temp).squeeze()
        #temp = temp.tolist()
        interpolated_data.append(temp)

    return interpolated_data

# 数据切分
#split the data size=buffer_size






# nomalize the data using pandas
```
